```python
#making changes to the data

#create a column that sums the total of all the pokemon stats
df['Total'] = df.iloc[:,4:10].sum(axis=1) 

#save the new file (as csv) - remove the index column
df.to_csv('modified.csv', index=False)
```

```python
#filtering data

#get pokenoms that are grass/poison
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]

#reset the index of the new dataframe, and drop the old indexes
new_df = new_df.reset_index(drop=True)

new_df
```

```python
#use regular expression to filter the data

#import regular expressions
import re

#get the fire or grass pokemons and be a case-insensitive search
new_df_2 = df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I,regex=True)]
new_df_2
```
